```python
import pandas as p
#p.read_csv() - read csv files 
data = p.read_csv('nato_phonetic_alphabet.csv')
#key:value - row.letter:row.code
nato_phonetic_dict = {row.letter:row.code for (index, row) in data.iterrows()}
#print(nato_phonetic_dict)
word = input('Enter a word: ').upper()
output = [nato_phonetic_dict[letter] for letter in word]
print(output)
```
